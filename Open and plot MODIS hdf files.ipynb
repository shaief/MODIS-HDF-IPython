# This IPython Notebook demonstrates how to work with HDF files:

#### Before starting:
The first cell does all the pre-work: The 1st line allows the plots to be inline with the code and not in external windows. Th 2nd and 3rd lines import relevant libraries (pyhdf for HDF reading and matplotlib for plotting).

```python
%pylab inline
from pyhdf.SD import SD, SDC
from matplotlib.pyplot import imshow
```

#### Working with the HDF file:
Now we're refering to the file itself. I created a variable named 'filename' that stores the file's name. Then i create an object using the pyhdf package, using the SD module, in a reading mode.

```python
filename = 'A2013336.L3m_DAY_CHL_chlor_a_1km.hdf'
hdf = SD(filename, SDC.READ)
hdf.datasets()
```

### Working with the data in the file:
Here i present some methods to work with the data stored in the file.

```python
l3m = hdf.select('l3m_data')  # select the specific data from the HDF file - here i chose 'l3m_data'
log_l3m = log(l3m.get())  # run logarithmic operation on the data
# l3m[0:10]  # slice a part of the original data
# log_l3m[0:10]  # slice a part of the logarithmic data
# autoscale(l3m)
```

### Plotting:
In order to plot, the matplotlib package has to be imported. In this notebook, this is done in the 3rd line in the 1st cell.

```python
fig = figure(figsize=(24,16), dpi=96)
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.imshow(log_l3m, vmin=-5, vmax=0, interpolation='bilinear')
title_text = filename.split('.') 
title('{} {}'.format(title_text[0],' '.join(title_text[1].split('_'))))
grid(True)
```
